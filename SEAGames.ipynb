```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt 

url = requests.get('https://www2.2019seagames.com/countries/')
file = BeautifulSoup(url.content, 'html.parser')

out = []

for i in file.find_all("em"):
    out.append(i.text)

out
```

```python
state_temp = []
state = []
total = []
gold = []
silver = []
bronze = []
year = []

count = 0
for i in file.find_all('em'):
    if count % 6 == 0 and count != 0:
        state.append(state_temp)
        state_temp = []
        state_temp.append(i.text)
        count += 1
    else:
        state_temp.append(i.text)
        count += 1

state = state[0:11]
state
```

```python
for i in range(len(state)):
    state[i][1] = state[i][1].replace('Total: ', '')
    state[i][2] = state[i][2].replace('Gold: ', '')
    state[i][3] = state[i][3].replace('Silver: ', '')
    state[i][4] = state[i][4].replace('Bronze: ', '')
    state[i][5] = state[i][5].replace('First Joined: ', '')
state
```

```python
goldMedals_raw = out[2::6]
goldMedals_raw
```

```python
goldMedals_raw1 = []
goldMedals = []

for i in range(len(goldMedals_raw)):
    temp = goldMedals_raw[i].split(" ")
    goldMedals_raw1.append(temp[1])
    
for j in range(len(goldMedals_raw1)):
    temp = int(goldMedals_raw1[j])
    goldMedals.append(temp)
    
goldMedals = goldMedals[:11]
goldMedals
```

```python
state[-2][-1] = '2003'
state[3][0] = 'LAO PDR'
state[0][0] = 'BRUNEI'
```

```python
import numpy as np
import pandas as pd
sea_malaysia = pd.DataFrame(state, columns=['State', 'Total Medal', 'Gold', 'Silver', 'Bronze', 'First Joined'])
```

```python
sea_malaysia[['Total Medal', 
              'Gold', 
              'Silver', 
              'Bronze', 
              'First Joined']] = sea_malaysia[['Total Medal', 
                                               'Gold', 
                                               'Silver', 
                                               'Bronze', 
                                               'First Joined']].astype(int)
```

```python
sea_malaysia
```

```python
state_2019 = []
state_temp_2019 = []
count = 0

url_2 = requests.get('https://rs.2019seagames.com/RS2019/mobiapp/MedalTally')
data_2 = BeautifulSoup(url_2.content, 'html.parser')
for i in data_2.find_all('small'):
    if count % 5 == 0 and count != 0:
        state_2019.append(state_temp_2019)
        state_temp_2019 = []
        state_temp_2019.append(i.text)
        count += 1
    else:
        state_temp_2019.append(i.text)
        count += 1
        
state_2019 = state_2019[1:]
state_2019.append(['TIMOR LESTE', '0', '0', '0', '0'])
state_2019.sort()
state_2019
```

```python
state_2019[0][0] = 'BRUNEI'
```

```python
state_2019
```

```python
sea_filipina = pd.DataFrame(state_2019, columns=['State',
                                                 'Gold', 
                                                 'Silver', 
                                                 'Bronze', 
                                                 'Total Medal'])
sea_filipina[['Gold', 
              'Silver', 
              'Bronze', 
              'Total Medal']] = sea_filipina[['Gold', 'Silver', 'Bronze', 'Total Medal']].astype(int)
sea_filipina
```
